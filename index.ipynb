## Transform to a Network Representation

Take the Pandas DataFrame and transform it into a graph representation via NetworkX. Make nodes the names of the various groups within the Nashville area. The edges should be the weights between these groups.

```python
#Your code here
groups = dict(zip(df2.group_id, df2.group_name))
```

```python
import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline

G = nx.Graph()

for i in df.index:
    group1 = groups[df['group1'][i]]
    group2 = groups[df['group2'][i]]
    weight = df['weight'][i]
    G.add_edge(group1, group2, weight=weight)

plt.figure()
nx.draw(G, with_labels=True, node_color="#1cf0c7", pos=nx.spring_layout(G, k=2, seed=5),
        node_size=40, alpha=.7, font_size=4);
```
